```
# Plot tipo 'barplot' : gráfico em barras com qtdade de pessoas que trabalhou em cada ano

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import sys

# Carregando arquivo csv:
fonte = "https://github.com/AI-Omar/CalcMax-Qtd-Trabalhador-Ano/blob/Teste/AnosInformados.csv?raw=true"
# Cria o DataFrame
df = pd.read_csv(fonte , ';')
#Atribui
df1 = df
# Cria coluna nova
df1.loc[:,'Pessoas'] = 0
# agrupa e conta
df1 = df.groupby(['all_year']).agg({'Pessoas': np.size}, ignore_index = False)
##altera label do eixo y
df1['Qtdade de Pessoas que trabalhou em cada ano'] = df1['Pessoas'] 
#criando uma figure, axes
fig, ax = plt.subplots(figsize=(8,6))
#altera label do eixo x
df1['Anos Trabalhados'] = df1.index
#criando o gráfico de barras
sns.barplot(x=df1['Anos Trabalhados'], y=df1['Qtdade de Pessoas que trabalhou em cada ano'], ax=ax, data=df1)
#otimizar espaço da figure
fig.tight_layout()
```
